```python
from bs4 import BeautifulSoup
url = 'http://books.toscrape.com/catalogue/page-{}.html'
import requests
pages = []
for x in range(1,51):
    myurl = url.format(x)
    page = requests.get(myurl)
    pages.append(page)
    
print(len(pages))


    
```

```python

for i in pages:
    soup = BeautifulSoup(i.content,"html.parser")
    titles = soup.find_all("h3")
    
    for x in titles:
        print(x.a.string)
    
```

```python
for i in pages:
    soup = BeautifulSoup(i.content,"html.parser")
    ratings = soup.find_all(attrs={"class":"star-rating"})
    for x in ratings:
        print(x['class'][1])
```

```python
def checker(a):
    if(a == "icon-ok"):
        return "In stock"
    return "Out of stock"
for i in pages:
    soup = BeautifulSoup(i.content,"html.parser")
    products = soup.find_all(attrs={"class":"product_pod"})
    for x in products:
        name = x.h3.a.text
        rating = x.find(attrs={"class":"star-rating"})
        product_attrs = x.find(attrs={"class":"product_price"})
        price = product_attrs.find(attrs={"class":"price_color"}).string
        availability = checker(product_attrs.find(attrs={"class":"availability"}).i['class'][0])
        print("Product Name :{} Product rating :{} Product price :{} Product Availability : {}".format(name,rating,price,availability))
        
```
